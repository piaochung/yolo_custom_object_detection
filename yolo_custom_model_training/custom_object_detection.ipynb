```python
%cd /
from glob import glob

train_img_list = glob('/content/drive/My Drive/yolo_custom_model_training/custom_dataset/train/images/*.jpg')
val_img_list = glob('/content/drive/My Drive/yolo_custom_model_training/custom_dataset/valid/images/*.jpg')

print(len(train_img_list), len(val_img_list))
```

```python
with open('/content/drive/My Drive/yolo_custom_model_training/custom_dataset/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/drive/My Drive/yolo_custom_model_training/custom_dataset/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml

with open('/content/drive/My Drive/yolo_custom_model_training/custom_dataset/data.yaml', 'r') as f:
  data = yaml.load(f)

data['train'] = '/content/drive/My Drive/yolo_custom_model_training/custom_dataset/train.txt'
data['val'] = '/content/drive/My Drive/yolo_custom_model_training/custom_dataset/val.txt'

with open('/content/drive/My Drive/yolo_custom_model_training/custom_dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
